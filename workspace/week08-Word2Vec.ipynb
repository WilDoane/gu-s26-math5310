```python
from gensim.models import Word2Vec

# Toy corpus (tokenized by whitespace)
sentences = [
    ["king", "queen", "man", "woman"],
    ["paris", "france", "rome", "italy"],
    ["man", "king"],
    ["woman", "queen"],
    ["paris", "capital", "france"],
    ["rome", "capital", "italy"]
]

# Train model
model = Word2Vec(
    sentences,
    vector_size=50,   # embedding dimension
    window=2,
    min_count=1,
    workers=1,
    seed=42
)
```

```python
vector = model.wv["king"]
print("Vector shape:", vector.shape)
print("First 5 values:", vector[:5])
```

```python
# cosine similarity
similarity = model.wv.similarity("king", "queen")
print("king ~ queen:", similarity)
```

```python
similarity = model.wv.similarity("king", "france")
print("king ~ france:", similarity)
```
